# **CLUSTERING & K-MEANS ALGORITHM**
##### Terdapat sebuah data yang menunjukan data penjualan dari beberapa brand kamera, total penjualan, dan rata-rata penjualan per harinya untuk kebutuhan analisis bisnis

```python
import matplotlib.pyplot as plt

rata_produksiperhari = [4000, 5000,100, 500, 60, 2000, 200, 40, 25, 20, 90, 10, 1000, 1100, 5, 220, 150]
total_jual = [3800, 4300, 30, 460, 35, 1340, 170, 15, 24, 15, 80, 4, 750, 1070 , 2, 200, 120]
rata_jual =  [3780, 4250, 20, 430, 28, 1290, 165, 10, 20, 10, 65, 3, 730, 1050 ,2, 195, 110]


plt.scatter(rata_produksiperhari, total_jual, rata_jual)
plt.show()
```

```python
from sklearn.cluster import KMeans

data = list(zip(rata_produksiperhari, total_jual, rata_jual))
inertias = []

for i in range(1,16):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,16), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)

plt.scatter(rata_produksiperhari, total_jual, rata_jual, c=kmeans.labels_)
plt.show()
```
